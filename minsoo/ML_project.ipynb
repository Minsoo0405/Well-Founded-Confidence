# 4) 다중선형 회귀분석

### 4-1) 데이터 이해 및 준비

```python
# 데이터 준비
train_df
```

```python
train_df.describe()
```

```python
# 데이터 분할
X = train_df.drop(['iid', 'match'], axis=1)
y = train_df['match']
```

```python
# MSE, RMSE, r2계수 함수 정의
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

def evaluate_score(y_test, y_pred):
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)


    print(f'mse - {mse:.4f}')
    print(f'rmse - {rmse: .4f}')
    print(f'r2 - {r2: .4f}')
    return
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=156)
```

### 4-2) 모델 구축

```python
# 모델 생성
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

### 4-3) 모델 학습

```python
# 모델 학습
reg = lr.fit(X_train, y_train)
reg
```

```python
coef = reg.coef_
inter = reg.intercept_

coef, inter
```

### 4-4) 모델 평가

```python
# 모델 평가
y_pred = reg.predict(X_test)

evaluate_score(y_test, y_pred)
```

# 5) 엘라스틱넷 모델

```python
# 5-1)모델 구축 및 학습(alpha=1, l1_ratio=0.5)
from sklearn.linear_model import ElasticNet
model_ElasticNet = ElasticNet(alpha=0.1, l1_ratio=0.5)
model_ElasticNet.fit(X_train, y_train)
```

```python
# 릿지 선형회귀 모델 평가(model_Ridge1, alpha=1)
# 선형 모델 평가
y_pred = model_ElasticNet.predict(X_test)
r2 = r2_score(y_test, y_pred)
print(r2)
```

```python
evaluate_score(y_test, y_pred)
```

### 5-2) LassoCV 모델을 통해 최적의 하이퍼파라미터 찾기

```python
#LassoCV()에 필요한 매개변수 설정
#0.01 ~ 9.91 까지 alpha_list  생성

cv = 5
alpha_list = np.arange(0.01, 10, 0.01)
max_iter = 1000
```

```python
from sklearn.linear_model import LassoCV

model_LassoCV = LassoCV(alphas=alpha_list, cv=5, max_iter=1000, n_jobs=-1, \
        random_state=42)
```

```python
model_LassoCV.fit(X_train, y_train)
```

```python
# alpha=8.89일때, 가장 좋은 score를 얻었다.
model_LassoCV.alpha_
```

```python
# 최적의 alpha를 가진 best_model로 다시 모델평가
from sklearn.linear_model import Lasso

best_model = Lasso(alpha=8.89)
best_model.fit(X_train, y_train)

y_pred = best_model.predict(X_test)
```

```python
# 하지만 LassoCV로 인해 최적의 alpha는 찾았기 때문에 새로운 모델을 작성할 필요가 없음
y_pred = model_LassoCV.predict(X_test)
r2_score(y_test, y_pred)
```

```python
evaluate_score(y_test, y_pred)
```

### 5-3) GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

model = ElasticNet()
param_grid = {
    'alpha' : [0.1, 1, 10, 100],
    'l1_ratio': [0.1, 0.3, 0.6, 0.9],
}

grid = GridSearchCV(model, param_grid=param_grid, cv=3)
```
