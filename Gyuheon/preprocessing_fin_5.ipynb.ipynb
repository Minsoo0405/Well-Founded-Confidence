```python
import pandas as pd
import numpy as np
```

```python
# 데이터 csv 불러오기
train = pd.read_csv('../data/train3.csv')
```

```python
# Unnamed: 0 column제거
train = train.iloc[:,1:]
# 소수점 2자리까지
train.match = round(train.match, 2)
```

```python
train.set_index('iid', inplace=True)
```

```python
train
```

```python
train.drop(['tmp_att', 'tmp_sin', 'tmp_int', 'tmp_fun', 'tmp_amb', 'tmp_sha'], axis=1, inplace=True)
```

```python
train.rename(columns={'gender': 'gender', 
                      'match': 'match', 
                      'attr3_1': 'attr_s', 
                      'sinc3_1': 'sinc_s', 
                      'intel3_1': 'intel_s', 
                      'fun3_1': 'fun_s', 
                      'amb3_1': 'amb_s', 
                      'tmp2_attr': 'attr_eval', 
                      'tmp2_sinc': 'sinc_eval', 
                      'tmp2_intel': 'intel_eval', 
                      'tmp2_fun': 'fun_eval', 
                      'tmp2_amb': 'amb_eval',
                      'tmp2_shar': 'shar_eval', 
                      'tmp2_like': 'like_eval'}, inplace=True)
```

```python
train.to_csv("train5.csv", index=False)
```
