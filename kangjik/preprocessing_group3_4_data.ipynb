```python
# 누락된 평가 확인
omission_list = list(set(g3.pid.unique()) - set(g3.iid.unique()))
omission_list
```

```python
# 누락된 평가를 가지고 있는 row의 인덱스 확인
df_omission_list = []
for omission in omission_list:
    df_omission_list.append(g3[g3.pid == omission].index)  

df_omission_list
    
```

```python
len(df_omission_list)
```

```python
# 누락된 평가를 가지고 있는 row 제거

for i in range(len(df_omission_list)):
    g3.drop(df_omission_list[i], axis=0, inplace=True)
```

```python
# iid 기준으로 그룹한 것을 iid_tmp에 저장
iid_tmp = g3.groupby('iid').mean()
iid_tmp
```

```python
# iid 기준 그룹화 df에서 결과를 보고 필요한 칼럼들 추출
# ['gender','match', 'attr3_1', 'sinc3_1', 'intel3_1','fun3_1','amb3_1']
iid_result_cols = ['gender','match', 'attr3_1', 'sinc3_1', 'intel3_1','fun3_1','amb3_1']
```

```python
# pid(상대방의 아이디)를 기준으로 그룹화
pid_tmp = g3.groupby('pid').mean()
pid_tmp
```

```python
# pid 기준 그룹화 df에서 결과에 필요한 칼럼들
pid_result_cols = 'pf_o_att	pf_o_sin	pf_o_int	pf_o_fun	pf_o_amb	pf_o_sha'.split('\t')
pid_result_cols += 'attr	sinc	intel	fun	amb	shar	like'.split()
pid_result_cols  += 'condtn, round, int_corr, age_o, age, attr1_1, sinc1_1, intel1_1, fun1_1, amb1_1, shar1_1, attr4_1, sinc4_1, intel4_1, fun4_1, amb4_1, shar4_1, attr2_1, sinc2_1, intel2_1, fun2_1, amb2_1, shar2_1, attr5_1, sinc5_1, intel5_1, fun5_1, amb5_1'.split(', ')
pid_tmp = pid_tmp[pid_result_cols]
```

```python
pid_tmp
```

```python
# pid 칼럼 명 변경
result_cols = ['tmp_att', 'tmp_sin', 'tmp_int', 'tmp_fun', 'tmp_amb', 'tmp_sha', 
               'tmp2_attr', 'tmp2_sinc', 'tmp2_intel', 'tmp2_fun', 'tmp2_amb', 'tmp2_shar', 'tmp2_like']
result_cols += 'condtn, round, int_corr, age_o, age, attr1_1, sinc1_1, intel1_1, fun1_1, amb1_1, shar1_1, attr4_1, sinc4_1, intel4_1, fun4_1, amb4_1, shar4_1, attr2_1, sinc2_1, intel2_1, fun2_1, amb2_1, shar2_1, attr5_1, sinc5_1, intel5_1, fun5_1, amb5_1'.split(', ')
pid_tmp.columns = result_cols 
pid_tmp
```

## 전처리한 데이터 프레임 만들기

```python
result = iid_tmp.copy()
```

```python
result[result_cols] = pid_tmp[result_cols]
```
