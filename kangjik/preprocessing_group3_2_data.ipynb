```python
g3 = g3[g3.iid != 414]
g3.info()
```

```python
# iid 기준으로 그룹한 것을 iid_tmp에 저장
iid_tmp = g3.groupby('iid').mean()
iid_tmp
```

```python
# iid 기준 그룹화 df에서 결과를 보고 필요한 칼럼들 추출
# ['gender','match', 'attr3_1', 'sinc3_1', 'intel3_1','fun3_1','amb3_1']
iid_result_cols = ['gender','match', 'attr3_1', 'sinc3_1', 'intel3_1','fun3_1','amb3_1']
```

```python
# pid(상대방의 아이디)를 기준으로 그룹화
pid_tmp = g3.groupby('pid').mean()
pid_tmp
```

```python
# pid 기준 그룹화 df에서 결과에 필요한 칼럼들
pid_result_cols = 'pf_o_att	pf_o_sin	pf_o_int	pf_o_fun	pf_o_amb	pf_o_sha'.split('\t')
pid_tmp = pid_tmp[pid_result_cols]
```

```python
# pid 칼럼 명 변경
result_cols = ['tmp_att', 'tmp_sin', 'tmp_int', 'tmp_fun', 'tmp_amb', 'tmp_sha']
pid_tmp.columns = result_cols 
pid_tmp
```

## 전처리한 데이터 프레임 만들기

```python
result = iid_tmp.copy()
```

```python
result[result_cols] = pid_tmp[result_cols]
```

```python
result
```

```python
result = result['gender	match	attr3_1	sinc3_1	intel3_1	fun3_1	amb3_1	tmp_att	tmp_sin	tmp_int	tmp_fun	tmp_amb tmp_sha'.split()]
```

```python
result.to_csv('../data/group3_2_fin.csv')
```
